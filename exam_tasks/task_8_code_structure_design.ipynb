## Part C (14)

Ultimately, Daedalus decides that the `Solver` class hierarchy *is* worthwhile.

- (2) Create the abstract base class `Solver`, assuming it would be placed inside the file `src/allocations/solver_base.py`. The class should:
    - have a class attribute `name` (a string) — the public name used by the dispatch (e.g. `"hungarian"`),
    - define an `__init__` method that takes no arguments,
    - define an abstract method `assign(self, matrix)` that subclasses must override.
    
    *Hint:* Use the `abc` module from the standard library.

- (8) Add (and write) two subclasses of `Solver`, one wrapping the existing `greedy` algorithm (call it `GreedySolver`) and one wrapping the existing `hungarian` algorithm (call it `HungarianSolver`). Each `assign(matrix)` method should:
    - delegate to the existing function in `allocations.solvers` rather than re-implementing the algorithm,
    - validate that the input matrix is non-empty before delegating, and raise `ValueError` if it is empty.
    
    Your code should make it clear (via class names, docstrings, or comments) which existing function each subclass corresponds to.

- (4) Replace the existing string-based dispatch in `solve(matrix, solver=...)` with a class-based dispatch that:
    1. maintains a registry of `Solver` subclasses keyed by their `name` class attribute,
    2. instantiates the matching subclass and returns the result of `solver.assign(matrix)`,
    3. raises a clear `ValueError` listing the available solver names when an unknown name is given.
    
    Your `solve` should preserve the existing public signature `solve(matrix, solver="hungarian")` and behaviour.

```python
# Solver base class (would live in src/allocations/solver_base.py)
from abc import ABC, abstractmethod


class Solver(ABC):
    """Abstract assignment solver."""

    name: str = ""

    def __init__(self):
        pass
   
        
        
    @abstractmethod
    def assign(self, matrix):
        pass
```

```python
# Concrete solvers wrapping the existing algorithms.
from allocations.solvers import greedy, hungarian


class GreedySolver(Solver):
    """Wraps allocations.solvers.greedy."""
    name = "greedy"
    
    
    
    def assign(self, matrix):
        
        if not matrix:
            raise ValueError("matrix is invalid")
        return self.greedy(matrix)
        
    


class HungarianSolver(Solver):
    """Wraps allocations.solvers.hungarian."""
    name = "hungarian"

    def assign(self, matrix):
        if not matrix:
            raise ValueError("Matrix must be non-zero")
        return hungarian(matrix)
        
     
```

```python
# Class-based dispatch replacing the existing solve()


_REGISTRY = {HungarianSolver.name: HungarianSolver, GreedySolver.name:GreedySolver}  # populated from Solver subclasses


def solve(matrix, solver="hungarian"):
    """Solve an assignment matrix using a named Solver subclass."""
    if solver in _REGISTRY.keys():
        result = _REGISTRY[solver]().assign(matrix)
        return result
    
    raise ValueError(f"Unknown solver {solver!r}. Available solvers: {list(_REGISTRY)}")

        

    
  
```
